```python
import matplotlib.pyplot as plt
import numpy as np

train_x = [-5, -4, -3, -2, -1, 0, 1, 2]
train_y = [2.8, 1.0, -1.0, -2.3, -2.7, -3.0, -3.2, -2.0]

plt.scatter(train_x, train_y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Train Dataset')
plt.grid(True)
plt.show()
```

```python
A = np.zeros((8, 3))
b = np.zeros((8, 1))

for i in range(8) :
    A[i][0] = train_x[i]**2
    A[i][1] = train_x[i]
    A[i][2] = 1
    b[i][0] = train_y[i]
    
print(A)
print(b)
```

```python
x = np.zeros((3, 1))
x = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), b)
print(x)
```

```python
input_x = np.linspace(-5, 5, 400)
a, b, c = x[0][0], x[1][0], x[2][0]
output_y = a * input_x**2 + b * input_x + c

plt.scatter(train_x, train_y)
plt.plot(input_x, output_y, label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of Quadratic Function')
plt.grid(True)
plt.show()
```
