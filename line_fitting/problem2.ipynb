```python
import matplotlib.pyplot as plt
import numpy as np

class Point :
    def __init__(self, x, y) :
        self.x = x
        self.y = y

points = []

user_input = input().split(' ')
for i in range(52) :
    if(i % 2 == 0) :
        x = int(user_input[i])
        tmp = Point(x, 0)
        points.append(tmp)
    else :
        y = int(user_input[i])
        y = 2268 - y
        points[(i - 1) // 2].y = y
    
for i in range(26) :
    print(points[i].x, points[i].y )
```

```python
train_x = []
train_y = []
for i in range(26) :
    train_x.append(points[i].x)
    train_y.append(points[i].y)

plt.scatter(train_x, train_y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Train Dataset')
plt.grid(True)
for i in range(26):
    plt.text(train_x[i], train_y[i], i + 1, ha='right', va='bottom')
plt.show()
```

```python
A = np.zeros((7, 2))
b = np.zeros((7, 1))

A[0][0] = points[0].x
A[0][1] = 1
b[0][0] = points[0].y

for i in range(1, 7) :
    A[i][0] = points[4 * i - 1].x
    A[i][1] = 1
    b[i][0] = points[4 * i - 1].y
    
x = np.zeros((2, 1))
x = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), b)

input_x = np.linspace(0, 3500, 400)
mv1, nv1 = x[0][0], x[1][0]
output_y = mv1 * input_x + nv1

plt.scatter(train_x, train_y)
plt.plot(input_x, output_y, label=f'y = {mv1}x + {nv1}')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph of first vertical line')
plt.grid(True)
plt.show()
```

```python
input_x = np.linspace(-1000, -250, 400)
output_y1 = mv1 * input_x + nv1
output_y2 = mv2 * input_x + nv2
output_y3 = mv3 * input_x + nv3
output_y4 = mv4 * input_x + nv4

class Line :
    def __init__(self, m, n) :
        self.m = m
        self.n = n
        
    def find_intersection(self, l) :
        x = (l.n - self.n) / (self.m - l.m)
        y = self.m * x + self.n
        intersection = Point(x, y)
        return intersection

lines = [Line(mv1, nv1), Line(mv2, nv2), Line(mv3, nv3), Line(mv4, nv4)]
intersections = []

for i in range(3) :
    for j in range(i + 1, 4) :
        intersections.append(lines[i].find_intersection(lines[j]))

intersection_x = []
intersection_y = []
        
for i in range(6) :
    intersection_x.append(intersections[i].x)
    intersection_y.append(intersections[i].y)
    print(intersections[i].x, intersections[i].y)
        
plt.scatter(intersection_x, intersection_y)
plt.plot(input_x, output_y1, label=f'y = {mv1}x + {nv1}')
plt.plot(input_x, output_y2, label=f'y = {mv2}x + {nv2}')
plt.plot(input_x, output_y3, label=f'y = {mv3}x + {nv3}')
plt.plot(input_x, output_y4, label=f'y = {mv4}x + {nv4}')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Intersection of the lines')
plt.grid(True)
plt.show()
```
